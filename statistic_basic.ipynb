#1. What is the difference between descriptive statistics and inferential statistics? Explain with examples.

Ans - In statistics, two major branches help us understand and analyze data: descriptive statistics and inferential statistics. Both are essential in data analysis, but they serve different purposes and are used in distinct scenarios. Descriptive statistics summarizes and presents the main features of a data set, while inferential statistics uses sample data to make broader generalizations about a larger population.

##1. Descriptive Statistics
Descriptive statistics deals with the methods of organizing, summarizing, and presenting data in an informative way. It involves the calculation and presentation of numerical measures that describe the characteristics of a data set. These statistics do not make predictions or inferences beyond the given data.

- Techniques used: Mean, Median, Mode, Standard Deviation, Percentages, Graphs, and Charts.

##Example:

If we collect the exam scores of 50 students in a class and find that the average score is 72, the median is 75, and the standard deviation is 8, these values describe only that particular class’s data.

##2.Inferential Statistics

Inferential statistics involves methods that allow us to make conclusions, predictions, or generalizations about a population based on information obtained from a sample. It goes beyond the available data and is widely used in research and decision-making.

- Techniques used: Hypothesis Testing, Confidence Intervals, Regression, t-test, Chi-square test, ANOVA.

##Example:
 If we take a random sample of 50 students from a university and calculate their average exam score as 72, we can use inferential statistics to estimate that the true average score of the entire university population lies between 70–74 with 95% confidence.

##Difference between Descriptive statistics and inferential statistics.

 Descriptive statistics refers to the methods used to summarize, organize, and present data in a meaningful way. It deals only with the data that has been collected and does not go beyond it. Common techniques include mean, median, mode, standard deviation, percentages, tables, graphs, and charts. For example, if we record the marks of 50 students in a class and find that the average score is 72, the median is 75, and the standard deviation is 8, then we are using descriptive statistics. Here, the focus is only on describing the performance of those 50 students without making any predictions about others.

On the other hand, inferential statistics involves drawing conclusions, predictions, or generalizations about a population based on a sample of data. It goes beyond the available dataset and helps in decision-making. The techniques used include hypothesis testing, confidence intervals, regression analysis, t-tests, chi-square tests, and ANOVA. For instance, if we take a random sample of 50 students from a university and find their average exam score to be 72, we can use inferential statistics to estimate that the true average score of all university students lies between 70 and 74 with a 95% confidence level. In this case, the findings are extended from the sample to the entire population.

#2. What is sampling in statistics? Explain the differences between random and stratified sampling.



Ans - Sampling is the process of selecting a subset of individuals, items, or observations (called a sample) from a larger set (called the population). Since studying an entire population is often difficult due to constraints of time, cost, and resources, statisticians rely on sampling to draw conclusions about the whole population. The key aim is that the sample should represent the characteristics of the population as closely as possible.

- Types of Sampling Methods

Two important methods of probability sampling are Random Sampling and Stratified Sampling.

##Random Sampling

In random sampling, every individual or element in the population has an equal and independent chance of being selected.

The selection is made purely by chance (e.g., by lottery method, random number tables, or computer-generated numbers).

- Example:

If a college has 1,000 students, and we want to select 100 for a survey, we can assign numbers to all students and use a random number generator to pick 100 names.

##Process:
A sample is selected from the entire population without any prior grouping or division.
##Purpose:
To get a representative view of the entire population as a whole.
##When to Use:
When you want to ensure that every single unit in the population has an equal chance of being chosen.
- Advantages:

Simple to apply and understand.

Eliminates selection bias.

Each member has an equal probability of selection.

- Disadvantages:

May not represent small but important subgroups of the population.

Requires a complete list of the population, which is not always available.



##2. Stratified Sampling

In stratified sampling, the population is divided into different groups or strata based on specific characteristics such as age, gender, income level, education, etc.

From each stratum, a random sample is drawn in proportion to its size in the population.

Example:

 In a school with 600 boys and 400 girls (total 1,000 students), if we want a sample of 100 students, we select 60 boys and 40 girls randomly to maintain the same proportion as in the population.
 ##Process:
After dividing the population into strata, a random sample is then drawn from each stratum.
##Purpose:
To ensure that specific subgroups are represented in the sample, which is especially useful when these subgroups have different characteristics and opinions.
##When to Use:
When the population has significant variability among different subgroups, and you need to make inferences about each subgroup as well as the whole population.

- Advantages:

Ensures all important groups are represented.

Provides more accurate and reliable results compared to simple random sampling.

Reduces sampling error.

- Disadvantages:

Requires prior knowledge of population characteristics.

More complex and time-consuming to organize.

#3. Define mean, median, and mode. Explain why these measures of central tendency are important.

Ans - Mean, median, and mode are fundamental statistical measures that describe the central tendency of a dataset by providing a single representative value. The mean is the arithmetic average, calculated by summing all values and dividing by the count; the median is the middle value in a sorted dataset; and the mode is the most frequently occurring value. These measures are important because they allow for the summarization of large datasets, simplify data analysis and interpretation, facilitate comparisons between different datasets, and provide a basis for informed decision-making by highlighting a typical or central value.  

##Mean:
 The mean is the average of a dataset. To calculate it, you add up all the numbers in the dataset and then divide by the total number of values.

 Mean =∑X/N

##Median:
 The median is the middle number in a dataset that has been ordered from smallest to largest. If there is an even number of data points, the median is the average of the two middle numbers.

 Median = (n+1)/2

##Mode:
 The mode is the value that appears most often in a dataset. A dataset can have one mode, multiple modes (bimodal or multimodal), or no mode at all.

 mode = l+[f1-f0)/(2f1-f0-f2)*h

#Why Measures of Central Tendency Are Important
- Summarization and Interpretation:

 They condense large, complex datasets into a single, easy-to-understand value that represents the typical or central characteristic of the data.
- Data Comparison:

These measures provide a basis for comparing different datasets. For example, you can compare the average test scores (mean) of two different classes to see which class performed better.
- Understanding Data Distribution:

By comparing the mean, median, and mode, you can gain insight into the shape of the data's distribution. In a perfectly symmetrical distribution, all three measures are equal, but in a skewed distribution (where data is pulled to one side), they will differ.
- Decision-Making:

In real-world applications, these measures support informed decisions. For instance, businesses use mode to determine the most popular product size, economists use median for income distribution, and educators use mean for average exam scores.
- Identification of Typical Values:

They help identify what is considered normal or typical within a dataset, which is useful for various fields, from public health to finance.

#4. Explain skewness and kurtosis. What does a positive skew imply about the data?


Ans - In statistics, analyzing the shape of a data distribution is crucial for understanding how data behaves. Two key characteristics that describe the shape of a distribution are skewness and kurtosis. These are numerical measures that go beyond central tendency and dispersion, helping analysts understand symmetry, tail behavior, and peakedness in data.
#1. Skewness

Skewness is a measure of the asymmetry of a probability distribution. It tells us whether the data values are symmetrically distributed around the mean or whether they are skewed (stretched) more to one side.

##Types of Skewness:

1. Positive Skew (Right-skewed):

- Tail on the right side is longer.

- Most values are concentrated on the left, with a few very large values.

- Mean > Median > Mode.

- Example: Income distribution (a few people earn much higher than the average).

2. Negative Skew (Left-skewed):

- Tail on the left side is longer.

- Most values are concentrated on the right, with a few very small values.

- Mean < Median < Mode.

- Example: Age at retirement (most retire around the same age, but a few retire very early).


#2. Kurtosis

Kurtosis measures the "tailedness" or peakedness of a distribution. It tells us whether data have heavy or light tails compared to a normal distribution.


##Types of Kurtosis:

- Mesokurtic (kurtosis ≈ 3): Normal distribution shape.

- Leptokurtic (kurtosis > 3): Distribution is more peaked with heavy tails → more extreme values (outliers).

- Platykurtic (kurtosis < 3): Distribution is flatter with lighter tails → fewer outliers.

#What Does a Positive Skew Imply?

- A positive skew (right-skewed distribution) implies the following:

- The majority of values lie below the mean.

- There are a few large values that pull the mean to the right.

- The mean is greater than the median, which is greater than the mode.

- Tail extends to the right, indicating extreme large values.

- Can distort averages and mislead interpretations if not accounted for.

Data:

Most numbers are small (2, 3, 4, 5).

One large number (15) pulls the tail to the right.

#5. Implement a Python program to compute the mean, median, and mode of a given list of numbers.
#numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]


```python
##5. Implement a Python program to compute the mean, median, and mode of a given list of numbers.
#numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]


import statistics as stats

# Given list
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Calculate mean, median, and mode
mean_val = stats.mean(numbers)
median_val = stats.median(numbers)
mode_val = stats.mode(numbers)   # mode() returns the most common value

# Display results
print("Numbers:", numbers)
print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
```

#6. Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
#list_x = [10, 20, 30, 40, 50]
#list_y = [15, 25, 35, 45, 60]


```python
 #6. : Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
#list_x = [10, 20, 30, 40, 50]
#list_y = [15, 25, 35, 45, 60]


import numpy as np

# Given lists
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert to numpy arrays
x = np.array(list_x)
y = np.array(list_y)

# Compute means
mean_x = np.mean(x)
mean_y = np.mean(y)

# Compute covariance manually
cov_xy = np.sum((x - mean_x) * (y - mean_y)) / len(x)

# Compute correlation coefficient
corr_xy = cov_xy / (np.std(x, ddof=0) * np.std(y, ddof=0))

# Display results
print("List X:", list_x)
print("List Y:", list_y)
print("Mean of X:", mean_x)
print("Mean of Y:", mean_y)
print("Covariance:", cov_xy)
print("Correlation Coefficient:", corr_xy)

```

#7. Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:
#data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]


```python
#7. Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:
#data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]


import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Draw boxplot
plt.figure(figsize=(6,4))
sns.boxplot(data, orient="h", color="skyblue")
plt.title("Boxplot of Data")
plt.xlabel("Values")
plt.show()

# --- Outlier Detection ---
Q1 = np.percentile(data, 25)   # First quartile (25th percentile)
Q3 = np.percentile(data, 75)   # Third quartile (75th percentile)
IQR = Q3 - Q1                  # Interquartile Range

# Outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
```

##Explanation of Result

The boxplot shows the median, quartiles (Q1, Q3), whiskers, and outliers.

Most of the data lies between 12 and 29, forming a fairly compact distribution.

35 is an outlier because it lies beyond the upper bound (31.0).

This means most values are clustered, but one unusually high value skews the data.

#8.  You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.
# Explain how you would use covariance and correlation to explore this relationship
# Write Python code to compute the correlation between the two lists:
#advertising_spend = [200, 250, 300, 400, 500]
#daily_sales = [2200, 2450, 2750, 3200, 4000]



Ans - Using Covariance and Correlation in Analysis

##Covariance:

Measures how two variables change together.

If covariance is positive, it means when advertising spend increases, daily sales also tend to increase.

If covariance is negative, then when advertising spend increases, daily sales decrease.

Limitation: The scale of measurement affects the covariance, so it’s hard to interpret the strength of the relationship directly.

##Correlation (Pearson’s r):

A standardized version of covariance.

It ranges between -1 and +1:

+1 → Perfect positive linear relationship.

-1 → Perfect negative linear relationship.

0 → No linear relationship.

Unlike covariance, correlation is dimensionless, making it easier to interpret and compare.


- Covariance will tell us if sales increase with ad spend.

- Correlation will tell us how strong that relationship is.




```python
import numpy as np

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert to numpy arrays
x = np.array(advertising_spend)
y = np.array(daily_sales)

# Compute covariance matrix
cov_matrix = np.cov(x, y, ddof=0)   # ddof=0 → population covariance
cov_xy = cov_matrix[0, 1]

# Compute correlation coefficient
corr_xy = np.corrcoef(x, y)[0, 1]

# Display results
print("Advertising Spend:", advertising_spend)
print("Daily Sales:", daily_sales)
print("Covariance:", cov_xy)
print("Correlation Coefficient:", corr_xy)
```

#9. Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.

# Write Python code to create a histogram using Matplotlib for the survey data:
#survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]



Ans - Summary Statistics and Visualizations to Use

To understand the distribution of survey scores (on a scale of 1-10),
Summary Statistics provide a quick, numerical overview of the data. For this dataset, you should consider:

Before launching a product, understanding the distribution of customer satisfaction scores (1–10 scale) is crucial:

Mean (Average): Shows the overall average satisfaction.

Median: The middle score; less affected by outliers.

Mode: The most frequently occurring score.

Standard Deviation: Measures how spread out the scores are (high std. dev. = customers vary a lot).

Minimum & Maximum: To see the range of responses.

##Visualizations:

Histogram: Shows how frequently each score occurs → highlights distribution shape.

Boxplot (optional): Good for spotting outliers and overall spread.

1. Measures of Central Tendency:

- Mean: The average score, which provides a single value representing the center of the data.

- Median: The middle value of the sorted scores. It is less affected by extreme values than the mean, offering a more robust measure of the typical score.

- Mode: The score that appears most frequently. This is useful for identifying the most common satisfaction level among customers.

2. Measures of Dispersion:

Standard Deviation: A measure of how spread out the scores are from the mean. A small standard deviation indicates that the scores are clustered closely around the mean, while a large standard deviation means they are more dispersed.


Visualizations help you see the shape and characteristics of the distribution. A histogram is the most effective visualization for this data, as it shows the frequency of each score and reveals the overall distribution pattern.

3. Histogram of Survey Scores


The histogram below, generated from your data, visualizes the frequency of each survey score. It clearly shows that the most frequent score is 7, with a high concentration of scores between 6 and 9.


```python
# --- Import Libraries ---
import matplotlib.pyplot as plt
import numpy as np
import statistics as stats
from IPython.display import Image

# --- Given Survey Data ---
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# --- Summary Statistics ---
mean_val = np.mean(survey_scores)
median_val = np.median(survey_scores)
mode_val = stats.mode(survey_scores)
std_dev = np.std(survey_scores, ddof=0)  # population std dev

print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
print("Standard Deviation:", std_dev)
print("Min:", min(survey_scores))
print("Max:", max(survey_scores))

# --- Histogram with Mean Line ---
plt.figure(figsize=(6,4))
plt.hist(survey_scores, bins=6, color="skyblue", edgecolor="black")
plt.axvline(mean_val, color="red", linestyle="--", linewidth=2, label=f"Mean = {mean_val:.2f}")
plt.title("Distribution of Customer Satisfaction Scores")
plt.xlabel("Survey Score (1-10)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Save histogram as PNG
plt.savefig("survey_histogram.png")

# Show plot inside Colab
plt.show()

# --- Display Saved Image in Colab ---
Image("survey_histogram.png")
```
